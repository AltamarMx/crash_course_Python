# Crea una nueva columna
$$Tsa = T_o +  \frac{I_s * a} { h} -RF$$

```python
h = 13.
a = 0.3
RF = 3.9
tmx['Tsa'] = tmx.To +  tmx.Ig * a /h - RF
tmx
```

```python
tmx[['Tsa','To']].loc['2018-03-10'].plot(figsize=(12,4))
```

## Problema, haz NaN la irradiancia para poder calcular el promedio mensual de Ig

```python
tmx_nan = tmx.copy()
tmx_nan.loc[tmx_nan.Ig<= 0.,'Ig'] = float('NaN')
tmx_nan
```

```python
tmx
```

```python
tmx_nan.resample('D').mean() # comparar Ig en tmx
```

```python
tmx.resample('D').mean()
```

```python
tmx.loc['2018-03-10'].resample('D').mean()
```

# Calcula la energ'ia Anual, mensual, diaria

$$ E = \sum I_g \Delta t $$

```python
energia_anual = tmx[['Ig','Ib']].resample('Y').sum()*(1/6)
energia_anual
```

```python
energia_anual.plot(kind='bar')
```

```python
energia_mensual = tmx[['Ig','Ib']].resample('M').sum()*(1/6)
energia_mensual
```

```python
energia_mensual.plot(kind='bar',figsize=(12,4))
```

```python
(tmx[['Ig','Ib']].loc['2018-03'].resample('D').sum()*1/6).plot(kind='bar',figsize=(20,3))
```
