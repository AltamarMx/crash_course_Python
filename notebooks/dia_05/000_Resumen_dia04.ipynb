1. Manejo de paquetes
    1. Instalar
    ```python
    pip install paquetito 
    ```
    1. Actualizar 
    ```python
    pip install --upgrade paquetito 
    ```
    1. Instalar version espec'ifica
    ```python
    pip install paquetito==1.2.3
    ```
    1. importar paquete
    ```python
    import pandas as pd   # donde pandas es el nombre y pd el alias 
    ```
    1. Desinstalar 
    ```python
    pip uninstall paquetito 
    ```
    1. Poner atenci'on a los mensajes, puede ser que se requiera --user
1. Lee y carga archivo de texto plano a df
    1. Carga archivo de texto plano sin especificar columna como 'indice
    ```python
    df = pd.read_csv('../dir1/dir2/archivo.csv') 
    ```
    1. Carga archivo de texto plano sin especificar columna como 'indice
    ```python
    file = '../dir1/dir2/archivo.csv'
    df = pd.read_csv(file) 
    ```
    1. Carga archivo de datos indicando columna que ser'a 'indice y la convierte a fecha
    ```python
    file = '../dir1/dir2/archivo.csv'
    df = pd.read_csv(file,index_col=0,parse_dates=True) 
    ```
    1. Visualiza r'apidamente 

    1. Visualiza r'apidamente con tama~no de figura

    1. Visualiza r'apidamente con un plot por columna
    ```python
    df.plot(subplots=True)
    ```
