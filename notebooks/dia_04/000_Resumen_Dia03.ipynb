```python
def operaciones():
    ''' Funcioon que saluda, limpia la casa, hace la tarea y juega con perritos'''
    print('Hola munde')
    print('Estoy limpiando la casa')
    perritos = ['lua','gwen','koba']
    for perrito in perritos:
        print('estoy paseando a ',perrito)
    print('Hago la tarea la final')
operaciones()
```

```python
def suma_lista(numeros):
    '''Suma los numeros contenidos en una lista'''
    suma = 0
    for numero in numeros:
        suma += numero
    return suma

numeritos = [1,2,3,4]
suma_lista(numeritos)
```

```python
def potencia(numero,potencia=2):
    return numero**potencia

potencia(3,potencia=4)
```
